```
col = ['has_superstructure_mud_mortar_stone', 'has_superstructure_stone_flag',
       'has_superstructure_cement_mortar_stone',
       'has_superstructure_mud_mortar_brick',
       'has_superstructure_cement_mortar_brick', 'has_superstructure_timber',
       'has_superstructure_bamboo', 'has_superstructure_rc_non_engineered',
       'has_superstructure_rc_engineered', 'has_superstructure_other',
       'has_secondary_use',
       'has_secondary_use_agriculture', 'has_secondary_use_hotel',
       'has_secondary_use_rental', 'has_secondary_use_institution',
       'has_secondary_use_school', 'has_secondary_use_industry',
       'has_secondary_use_health_post', 'has_secondary_use_gov_office',
       'has_secondary_use_use_police', 'has_secondary_use_other']
```

```
for feat in col:
  final[feat] = final[feat].astype(str)
final.head()
```

```
feat_cat = [feature for feature in train.columns if final[feature].dtype == 'O']
feat_cat
```

```
for feature in feat_cat:
    print('The feature is {} and number of categories are {}'.format(feature,len(final[feature].unique())))
```

```
final.head()
```

```
for feat in feat_cat:
  Mean_encode = final.groupby([feat])['damage_grade'].mean().to_dict()  
  final[feat]=final[feat].map(Mean_encode)
final.head()
```

```
final['geo_level_1_id'] = final['geo_level_1_id']/final['area_percentage']
final['geo_level_2_id'] = final['geo_level_2_id']/final['area_percentage']
final['geo_level_3_id'] = final['geo_level_3_id']/final['area_percentage']
final.head()
```

```
col = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id'] 	
```

```
for feat in col:
  Mean_encode = final.groupby([feat])['damage_grade'].mean().to_dict()  
  final[feat]=final[feat].map(Mean_encode)
final.head()
```

```
df=final.loc[:,final.columns!='damage_grade']
df['damage_grade']=final['damage_grade']
df.head()
```

```
final=df
final.head()
```

```
final = final.drop(['building_id', 'area_percentage'], axis = 1)
final.head()
```

```
final.isnull().sum()
```

```
col = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id']
```

```
def impute_nan(final, feat):
    m=final[feat].mean()
    final[feat]=final[feat].fillna(m)

for i in col:
    impute_nan(final, i)
final.head()
```

```
final.isnull().sum()
```

```
scale_data=[feature for feature in final.columns if feature not in ['damage_grade']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
final[scale_data]=scaler.fit_transform(final[scale_data])
final.head()
```
